```python
import pandas as pd
import json
import random
```

```python
f = open("./imdb_yelp/yelp_training_set/yelp_training_set_business.json","r")
```

```python
lines = f.readlines() 
prd_names = []
```

```python
for line in lines:
    temp = json.loads(line)
    prd_names.append(temp["name"])
```

```python
prd_names = list(set(prd_names))
prd_names
```

```python
len(prd_names)
```

```python
f = open("./imdb_yelp/yelp_training_set/vocab.txt", "r", encoding="utf8")
lines = f.readlines()
vocab_names =[]
for line in lines:
    vocab_names.append(line.strip())
```

```python
vocab_names = list(set(vocab_names))
len(vocab_names)
```

```python
intersection = []
```

```python
count = 0
for name in prd_names:
    split_name = name.split()
    found = True
    for split in split_name:
        if split.lower() in vocab_names:
            continue
        else:
            found = False
            break
            
    if found:
        intersection.append(name)
        count = count + 1        
```

```python
count
```

```python
intersection
```

```python
names = random.sample(intersection, k=1633)
len(names)
```

```python
f = open("./imdb_yelp/yelp_training_set/yelp_training_set_user.json","r")
lines = f.readlines() 
```

```python
user_names = []
for line in lines:
    temp = json.loads(line)
    user_names.append(temp["name"])
user_names = list(set(user_names))
```

```python
user_names
```

```python
intersection_user = []
count = 0
for name in user_names:
    split_name = name.split()
    found = True
    for split in split_name:
        if split.lower() in vocab_names:
            continue
        else:
            found = False
            break
            
    if found:
        intersection_user.append(name)
        count = count + 1        
```

```python
count
```

```python
intersection_user
```

```python
user_names = random.sample(intersection_user, k=1631)
len(user_names)
```

```python
train = pd.read_csv("./imdb_yelp/yelp-2013-seg-20-20.train.ss", sep="\t\t", names=["user_id", "product_id", "rating", "text"])
dev = pd.read_csv("./imdb_yelp/yelp-2013-seg-20-20.dev.ss", sep="\t\t", names=["user_id", "product_id", "rating", "text"])
test = pd.read_csv("./imdb_yelp/yelp-2013-seg-20-20.test.ss", sep="\t\t", names=["user_id", "product_id", "rating", "text"])
```

```python
train["text"] = train["text"].apply(lambda x: x.replace("<sssss>", ""))
dev["text"] = dev["text"].apply(lambda x: x.replace("<sssss>", ""))
test["text"] = test["text"].apply(lambda x: x.replace("<sssss>", ""))
```

```python
product_id_train = train["product_id"].unique()
product_id_dev = dev["product_id"].unique()
product_id_test = test["product_id"].unique()
print(len(product_id_train))
print(len(product_id_dev))
print(len(product_id_test))
```

```python
user_id_train = train["user_id"].unique()
user_id_dev = dev["user_id"].unique()
user_id_test = test["user_id"].unique()
print(len(user_id_train))
print(len(user_id_dev))
print(len(user_id_test))
```

```python
uid_to_name = {}
c = 0
for i in user_id_train:
    uid_to_name[i] = user_names[c]
    c = c + 1
```

```python
id_to_name = {}
c = 0
for i in product_id_train:
    id_to_name[i] = names[c]
    c = c + 1
```

```python
train["business"] = train["product_id"].apply(lambda x: id_to_name[x])
dev["business"] = dev["product_id"].apply(lambda x: id_to_name[x])
test["business"] = test["product_id"].apply(lambda x: id_to_name[x])
```

```python
train["user"] = train["user_id"].apply(lambda x: uid_to_name[x])
dev["user"] = dev["user_id"].apply(lambda x: uid_to_name[x])
test["user"] = test["user_id"].apply(lambda x: uid_to_name[x])
```

```python
train_df = pd.DataFrame({"id":range(train.shape[0]), "text":train["text"], "label":train["rating"]-1, "business":train["business"], "user":train["user"]})
dev_df = pd.DataFrame({"id":range(dev.shape[0]), "text":dev["text"], "label":dev["rating"]-1,  "business":dev["business"], "user":dev["user"]})
test_df = pd.DataFrame({"id":range(test.shape[0]), "text":test["text"], "label":test["rating"]-1,  "business":test["business"], "user":test["user"]})
```

```python
train_df.to_csv("./yelp2013final/train.csv", sep=",", index=False)
dev_df.to_csv("./yelp2013final/dev.csv", sep=",", index=False)
test_df.to_csv("./yelp2013final/test.csv", sep=",", index=False)
```

```python
train_df["text"] = "#" + train_df["user"] + "# $" + train_df["business"] + "$ " + train_df["text"] 
dev_df["text"] = "#" + dev_df["user"] + "# $" + dev_df["business"] + "$ " + dev_df["text"] 
test_df["text"] = "#" + test_df["user"] + "# $" + test_df["business"] + "$ " + test_df["text"] 
```
